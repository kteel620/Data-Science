##### Guided Practice: Logit Function and Odds

```python
def logit_func(odds):
    # uses a float (odds) and returns back the log odds (logit)
    return math.log(odds)

def sigmoid_func(logit):
    # uses a float (logit) and returns back the probability
    return math.exp(logit)

odds_set = [
    5./1.0,
    20./1.0,
    1.1/1.0,
    1.8/1.0,
    1.6/1.0
]
```

```python
#test
from math import log
print(math.log(math.e))
```

```python
new_log_odds = []
for i in odds_set:
    new_log_odds.append(logit_func(i))
print(new_log_odds)
```

```python
new_sig_odds = []
for i in new_log_odds:
    new_sig_odds.append(sigmoid_func(i))
print(new_sig_odds)
    
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
# Statsmodels logistic regression is sm.Logit
import statsmodels.api as sm
```

```python
titanic = pd.read_csv('../../assets/dataset/titanic.csv')
```

```python
titanic
```

```python
titanic['Age'] = titanic.groupby(["Sex", 'Pclass']).Age.transform(lambda x: x.fillna(x.mean()))
titanic['had_parents'] = titanic.Parch.apply(lambda x: 1 if x > 0 else 0)
titanic['had_siblings'] = titanic.SibSp.apply(lambda x: 1 if x > 0 else 0)
titanic['is_female'] = titanic.Sex.apply(lambda x: 1 if x == 'female' else 0)
titanic.head()
```

```python
B = pd.get_dummies(titanic["Pclass"], prefix = "Class")
titanic = pd.concat([titanic, B], axis = 1)
titanic.head()
```

```python
titanic["Pclass"].value_counts()
```

```python
X = titanic[['is_female',"Class_1", "Class_2"]]

y = titanic['Survived']

lm = sm.Logit(y, X)
result = lm.fit()

result.summary()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
lm = LogisticRegression() 
```

```python
lm.fit(X_train, y_train)
```

```python
predicted_values = lm.predict(X_test)
```

```python
import sklearn.metrics
```

```python
conf_matrix = sklearn.metrics.confusion_matrix(y_test, predicted_values)
```
